```python
co2 = pd.read_csv('data/CO2concentration.csv')
```

```python
co2['date'] = pd.to_datetime((co2.Year*10000+co2.Month*100+1).apply(str),format='%Y%m%d')
```

```python
plt.figure(figsize = (12,5))
plt.plot(co2.date, co2.CO2)
plt.plot(co2.date, pd.ewma(co2.CO2,com=7))
plt.title('CO2 Emissions')
plt.ylabel('parts-per-million (ppm)')
plt.xlabel('year')
plt.legend(labels=['data', 'smoooooothed data'], loc=2)
```
